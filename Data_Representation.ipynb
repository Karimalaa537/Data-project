READING FILE

```python
data=pd.read_csv('responses.csv')
data.head(5)
```

BAR CHART

```python
x = data['On a scale of 1 to 5, how influential is social media in shaping your awareness of local clothing brands? (1 being not influential, 5 being extremely influential)']
plt.figure(figsize=(8, 6))
plt.hist(x, bins=5, color='#ADD8E6', edgecolor='black')
plt.xlabel('Influence of Social Media on Awareness')
plt.ylabel('Frequency')
plt.title('Distribution of Influence of Social Media on Awareness')
plt.show()
```

PIE CHART

```python
effect_counts = data['what type of effect has social media had on your purchasing decisions related to local clothing brands'].value_counts()
plt.pie(effect_counts, labels=effect_counts.index, autopct='%1.1f%%')
plt.title('Effect of Social Media on Purchasing Decisions')
plt.axis('equal')
plt.show()
```
